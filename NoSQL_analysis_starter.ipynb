### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
pprint(db.establishments.find_one({"BusinessName": "Penang Flavours"}))
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "LocalAuthorityName": {"$regex": "london", "$options": "i" },
"RatingValue":{"$gte": 4}}

# Use count_documents to display the number of documents in the result
results_count = db.establishments.count_documents(query)

# Display the first document in the results using pprint
print(f'There are {results_count} documents with the Local Authority of London and rating value greater than or equal to 4')
```

```python
# Convert the result to a Pandas DataFrame
london_authority_df = pd.DataFrame(db.establishments.find(query))

# Display the number of rows in the DataFrame
print(f'There are {len(london_authority_df)} rows in the London Authority Data Frame')

# Display the first 10 rows of the DataFrame
london_authority_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
degree_search = 0.01
max_lat = db.establishments.find_one({"BusinessName": "Penang Flavours"})['geocode']['latitude'] + degree_search
min_lat = db.establishments.find_one({"BusinessName": "Penang Flavours"})['geocode']['latitude'] - degree_search
max_long = db.establishments.find_one({"BusinessName": "Penang Flavours"})['geocode']['longitude'] + degree_search
min_long = db.establishments.find_one({"BusinessName": "Penang Flavours"})['geocode']['longitude'] - degree_search
```

```python
# Rating value must equal 5
# Sort by hygiene score

query = {"RatingValue": 5, 
        "geocode.latitude": {"$lte": max_lat, "$gte": min_lat},
        "geocode.longitude": {"$lte": max_long, "$gte": min_long}
         }

sort =  [("scores.Hygiene", -1)]

limit = 5

#get the results
results = db.establishments.find(query).sort(sort).limit(limit)

# Print the results
pprint(list(results))
```

```python
# Convert result to Pandas DataFrame
top5_df = pd.DataFrame(list(results))
top5_df
```
